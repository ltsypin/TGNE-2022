```python
annotations_none = pd.read_csv('./TGNE/eggnog/2024_none_eggnog_compiled.annotations', comment='#', delimiter='\t')
annotations_none = annotations_none.sort_values(by='query')
annotations_none.reset_index(drop=True, inplace=True)

annotations_realign = pd.read_csv('./TGNE/eggnog/2024_realign_eggnog_compiled.annotations', comment='#', delimiter='\t')
annotations_realign = annotations_realign.sort_values(by='query')
annotations_realign.reset_index(drop=True, inplace=True)

annotations_denovo = pd.read_csv('./TGNE/eggnog/2024_denovo_eggnog_compiled.annotations', comment='#', delimiter='\t')
annotations_denovo = annotations_denovo.sort_values(by='query')
annotations_denovo.reset_index(drop=True, inplace=True)
```
