```python
import pandas as pd
import numpy as np
```

```python
def max_aggregate_df(df: pd.DataFrame, aggregate_col: str, max_col: str):    
    max_df = None

    cols = list(df.columns)

    for val in df[aggregate_col].unique():

        curr_df_raw = df.loc[df[aggregate_col] == val]

        max_value = curr_df_raw[max_col].max()

        curr_df = curr_df_raw.loc[curr_df_raw[max_col] == max_value]

        curr_df = pd.DataFrame({col: [','.join(list([str(_) for _ in curr_df[col].unique()]))] for col in cols})

        if max_df is None:
            max_df = curr_df
            continue

        max_df = pd.concat((max_df, curr_df), ignore_index=True)

    return max_df
```

```python
raw_df = pd.read_csv('./module_comparison.csv')
raw_df.head()
```

```python
rna_seq_df = max_aggregate_df(raw_df, 'rna_seq_mod', 'rna_seq_overlap_fraction')
```

```python
rna_seq_df['rna_seq_overlap_fraction'] = pd.to_numeric(rna_seq_df['rna_seq_overlap_fraction'], errors='coerce')
rna_seq_df.sort_values(by='rna_seq_overlap_fraction', ascending=False)
```

```python
microarray_df = max_aggregate_df(raw_df, 'microarray_mod', 'microarray_overlap_fraction')
```

```python
microarray_df['microarray_overlap_fraction'] = pd.to_numeric(microarray_df['microarray_overlap_fraction'], errors='coerce')
microarray_df.sort_values(by='microarray_overlap_fraction', ascending=False)
```
