```python
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score

# ARI
ari = adjusted_rand_score(subset_labels_1, subset_labels_2)
print(f"Adjusted Rand Index: {ari}")

# NMI
nmi = adjusted_mutual_info_score(subset_labels_1, subset_labels_2)
print(f"Normalized Mutual Information: {nmi}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(subset_labels_1, subset_labels_2)

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square statistic: {chi2}, p-value: {p}")
```

```python
import pandas as pd

# Example contingency table
contingency_table = pd.crosstab(subset_labels_1, subset_labels_2)

# Find the maximum value in the contingency table
max_value = contingency_table.max().max()

# Find the coordinates (x and y axis labels) of the maximum value
max_coords = contingency_table.unstack().idxmax()

# Print the results
print(f"Maximum value: {max_value}")
print(f"x label: {max_coords[1]}")
print(f"y label: {max_coords[0]}")
```
