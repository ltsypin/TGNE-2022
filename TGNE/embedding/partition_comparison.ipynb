# IMPORTS

```python
import pandas as pd
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score
```

# IMPORT TWO PARTITION DATASETS FOR COMPARISION

```python
df1 = pd.read_csv('./test_nn3_leiden_label_df_round_1.csv')
print(df1.shape)
df1.head()
```

```python
df2 = pd.read_csv('./rna_seq_label_df_round_1.csv')
print(df2.shape)
df2.head()
```

# FIND THE GENES INCLUDED IN BOTH DATASETS

```python
df1_genes = set(df1['TTHERM_ID'].values)
df1_genes
```

```python
df2_genes = set(df2['TTHERM_ID'].values)
df2_genes
```

```python
intersecting_genes = set.intersection(df1_genes, df2_genes)
intersecting_genes
```

# ISOLATE THE ROWS IN EACH DF WITH INTERSECTING GENES

```python
df1_filtered = df1.loc[df1['TTHERM_ID'].isin(intersecting_genes)]
print(df1_filtered.shape)
df1_filtered.head()
```

```python
df2_filtered = df2.loc[df2['TTHERM_ID'].isin(intersecting_genes)]
print(df2_filtered.shape)
df2_filtered.head()
```

# SORT THE FILTERED DFS

```python
df1_filtered_sorted = df1_filtered.sort_values(by='TTHERM_ID')
df1_filtered_sorted.head()
```

```python
df2_filtered_sorted = df2_filtered.sort_values(by='TTHERM_ID')
df2_filtered_sorted.head()
```

```python
df1_filtered_sorted['label'].values
```

```python
df2_filtered_sorted['label'].values
```

# COMPUTE THE COMPARISON METRICS

```python
adjusted_rand_score(df1_filtered_sorted['label'].values, df2_filtered_sorted['label'].values)
```

```python
adjusted_mutual_info_score(df1_filtered_sorted['label'].values, df2_filtered_sorted['label'].values)
```
