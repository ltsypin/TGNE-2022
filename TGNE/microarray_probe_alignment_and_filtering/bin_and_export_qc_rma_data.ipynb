```python
import pandas as pd
from json import load
```

```python
qc_rma_signal_data_df = pd.read_csv('../microarray_QC/QC_probe_rma_values.csv')
qc_rma_signal_data_df.head()
```

```python
qc_rma_signal_data_df.rename(columns={'Unnamed: 0': 'TTHERM_ID'}, inplace=True)
qc_rma_signal_data_df.rename(columns={c: c.split('.')[0] for c in qc_rma_signal_data_df.columns}, inplace=True)
qc_rma_signal_data_df.head()
```

```python
len(qc_rma_signal_data_df['TTHERM_ID'].values)
```

```python
with open('../../new_raw_data/microarray_accessions_all.json', 'r') as f:
    # Including single REP measurement for C-15m (GSM656231) even though there are no replicates for it
    # in order to replicate the 2011 analysis
    # Format: keys are the physiological phase; values are the geo accessions for each microarray
    all_geo = load(f)
all_geo
```

```python
col_names = []

for key in all_geo:
    for val in all_geo[key]:
        col_names.append(f'{key}_{val}')
```

```python
ordered_columns = [c for c in col_names if c in qc_rma_signal_data_df.columns[1:]]
ordered_columns
```

```python
tidy_column_order_list = ['TTHERM_ID'] + ordered_columns
tidy_column_order_list
```

```python
len(tidy_column_order_list)
```

```python
len(list(qc_rma_signal_data_df.columns)) 
```

```python
tidy_qc_rma_signal_data_df = qc_rma_signal_data_df.loc[:,tidy_column_order_list]
tidy_qc_rma_signal_data_df.head()
```

```python
tidy_qc_rma_signal_data_df.shape
```

```python
duplicates = tidy_qc_rma_signal_data_df['TTHERM_ID'].duplicated()
tidy_qc_rma_signal_data_df[duplicates]
```

```python
growth = ['Ll', 'Lm', 'Lh']
starvation = ['S0', 'S3', 'S6', 'S9', 'S12', 'S15', 'S24']
conjugation = ['C0', 'C15m', 'C2', 'C4', 'C6', 'C8', 'C10', 'C12', 'C14', 'C16', 'C18']
vegetative = growth + starvation

grow_cols = [c for c in tidy_qc_rma_signal_data_df.columns if c.split('_')[0] in growth]
starve_cols = [c for c in tidy_qc_rma_signal_data_df.columns if c.split('_')[0] in starvation]
sex_cols = [c for c in tidy_qc_rma_signal_data_df.columns if c.split('_')[0] in conjugation]
veg_cols = [c for c in tidy_qc_rma_signal_data_df.columns if c.split('_')[0] in vegetative]
```

```python
tidy_qc_rma_signal_data_df_grow = tidy_qc_rma_signal_data_df[['TTHERM_ID'] + grow_cols]
tidy_qc_rma_signal_data_df_starve = tidy_qc_rma_signal_data_df[['TTHERM_ID'] + starve_cols]
tidy_qc_rma_signal_data_df_sex = tidy_qc_rma_signal_data_df[['TTHERM_ID'] + sex_cols]
tidy_qc_rma_signal_data_df_veg = tidy_qc_rma_signal_data_df[['TTHERM_ID'] + veg_cols]
```

```python
tidy_qc_rma_signal_data_df.to_csv('./agg_tidy_2021aligned_qc_rma_expression_full.csv', index=False)
tidy_qc_rma_signal_data_df_grow.to_csv('./agg_tidy_2021aligned_qc_rma_expression_grow.csv', index=False)
tidy_qc_rma_signal_data_df_starve.to_csv('./agg_tidy_2021aligned_qc_rma_expression_starve.csv', index=False)
tidy_qc_rma_signal_data_df_veg.to_csv('./agg_tidy_2021aligned_qc_rma_expression_veg.csv', index=False)
tidy_qc_rma_signal_data_df_sex.to_csv('./agg_tidy_2021aligned_qc_rma_expression_sex.csv', index=False)
```
