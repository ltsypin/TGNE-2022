```python
import pandas as pd
import numpy as np

np.random.seed(393)
```

```python
all_good_filtered_df = pd.read_csv('../microarray_probe_alignment_and_filtering/allgood_filt_agg_tidy_2021aligned_qc_rma_expression_full.csv')
all_good_filtered_df.head()
```

```python
row_before = all_good_filtered_df[all_good_filtered_df['Unnamed: 0'] == 'TTHERM_000000045'].loc[0].to_list()
row_before
```

```python
columns_to_shuffle = all_good_filtered_df.columns[1:]
print(len(columns_to_shuffle))
columns_to_shuffle
```

```python
def shuffle_row(row):
    shuffled_row = row.values.copy()
    np.random.shuffle(shuffled_row)
    return pd.Series(shuffled_row, index=row.index)
```

```python
all_good_filtered_df[columns_to_shuffle] = all_good_filtered_df[columns_to_shuffle].apply(shuffle_row, axis=1)
```

```python
all_good_filtered_df.head()
```

### Sanity check to ensure only values in the rows were shuffled:

```python
row_after = all_good_filtered_df[all_good_filtered_df['Unnamed: 0'] == 'TTHERM_000000045'].loc[0].to_list()
row_after
```

```python
len(row_before) == len(row_after)
```

```python
for value in row_before:
    if value not in row_after:
        print(value)
```

```python
for idx in range(len(row_before)):
    if row_before[idx] != row_after[idx]:
        print('DIFFERENT:', row_before[idx], 'IS NOT', row_after[idx])
```
