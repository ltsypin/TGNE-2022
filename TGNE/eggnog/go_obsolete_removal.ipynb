```python
modified_annotation = complete_annotation.copy(deep=True)

obsolete_idxs = []

for idx, row in modified_annotation.iterrows():
    if row['GOs'] == '-':
        continue
    
    go_terms = row['GOs'].split(',')

    go_terms_no_obsolete = []

    for gt in go_terms:
        name, definition, obsolete = annotation_info.get_GO_info(gt)
        if obsolete == 'False':
            go_terms_no_obsolete.append(gt)
        else:
            if idx not in obsolete_idxs:
                obsolete_idxs.append(idx)

    if len(go_terms_no_obsolete) > 0:
        row['GOs'] = ','.join(go_terms_no_obsolete)
    else:
        row['GOs'] = '-'
```
