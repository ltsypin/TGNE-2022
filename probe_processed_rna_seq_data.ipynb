```python
df_rna_seq_data_ttherm = df_rna_seq_data.loc[df_rna_seq_data['gene'].str.startswith('TTHERM')]
df_rna_seq_data_ttherm
```

```python
df_rna_seq_data_not_ttherm = df_rna_seq_data.loc[~df_rna_seq_data['gene'].str.startswith('TTHERM')]
df_rna_seq_data_not_ttherm
```

These are the 3244/26000 genes that they determined to be cell cycle regulated in the paper.

```python
import requests
from bs4 import BeautifulSoup

dict_alias_to_ttherm_id = {}

url_begin = "https://tet.ciliate.org/search.php?gene_name="
url_ends = list(df_rna_seq_data_not_ttherm['gene'].values)

print(url_ends)

for url_end in url_ends:

    url = f'{url_begin}{url_end}'

    response = requests.get(url)

    if response.status_code == 200:
        html_content = response.content
        soup = BeautifulSoup(html_content, 'html.parser')

        table_element = soup.find('table', class_='main')

        if table_element:
            # Find the <a> element with class "geneName" within the table
            gene_name_element = table_element.find('a', class_='geneName')

            if gene_name_element:
                ttherm_id = gene_name_element.text.strip()
                dict_alias_to_ttherm_id[url_end] = ttherm_id

            else:
                print(f"No geneName element found for {url_end}")
        else:
            print(f"No table element found for {url_end}")
    else:
        print(f"Failed to retrieve content. Status code: {response.status_code}")
```
