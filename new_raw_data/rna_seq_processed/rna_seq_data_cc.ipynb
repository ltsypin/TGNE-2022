```python
import pandas as pd
import numpy as np
```

```python
df_manual = pd.read_csv('./tpm_manual.csv')
df_manual.head()
```

```python
df_kallisto = pd.read_csv('./tpm_kallisto.csv')
df_kallisto.head()
```

```python
df_manual.shape
```

```python
df_kallisto.shape
```

```python
manual_genes = set(df_manual['TTHERM_ID'].values)
len(manual_genes)
```

```python
kallisto_genes = set(df_kallisto['TTHERM_ID'].values)
len(kallisto_genes)
```

```python
len(set.intersection(manual_genes, kallisto_genes))
```

```python
genes_intersection = list(set.intersection(manual_genes, kallisto_genes))
genes_intersection
```

```python
len(manual_genes-kallisto_genes)
```

```python
len(kallisto_genes-manual_genes)
```

```python
df_kallisto_intersect = df_kallisto.loc[df_kallisto['TTHERM_ID'].isin(genes_intersection), :]
df_kallisto_intersect.reset_index(inplace=True, drop=True)
df_kallisto_intersect
```

```python
df_manual_intersect = df_manual.loc[df_manual['TTHERM_ID'].isin(genes_intersection), :]
df_manual_intersect.reset_index(inplace=True, drop=True)
df_manual_intersect
```

```python
df_kallisto_intersect.shape
```

```python
df_manual_intersect.shape
```

```python
df_kallisto_intersect.loc[:, df_kallisto_intersect.columns[1:]]
```

```python
df_manual_intersect.loc[:, df_manual_intersect.columns[1:]]
```

```python
df_intersect_divided = df_kallisto_intersect.loc[:, df_kallisto_intersect.columns[1:]] / df_manual_intersect.loc[:, df_manual_intersect.columns[1:]]
df_intersect_divided.head()
```

```python
float_cols = df_intersect_divided.columns
```

```python
list(df_manual_intersect['TTHERM_ID'].values) == list(df_kallisto_intersect['TTHERM_ID'].values)
```

```python
df_intersect_divided['TTHERM_ID'] = df_manual_intersect['TTHERM_ID'].values
df_intersect_divided.head()
```

```python
df_intersect_divided['mean'] = df_intersect_divided[float_cols].mean(axis=1)
df_intersect_divided['sd'] = df_intersect_divided[float_cols].std(axis=1)
df_intersect_divided.head()
```

```python
df_intersect_divided.shape
```

```python
df_intersect_divided.loc[df_intersect_divided['sd'] < 3]
```

```python
df_intersect_divided['sd'].min()
```

```python
df_intersect_divided['sd'].mean()
```

```python
df_intersect_divided['sd'].median()
```
