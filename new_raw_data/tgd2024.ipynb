# NEW 2024 MODEL (TTHERM_IDs) (MAY 15)

```python
with open('./tgd2024_may15/Tetrahymena_Genome_annotation_V2024_CDS.fasta', 'r') as f:
    cds_may15_records = list(SeqIO.parse(f, 'fasta'))

cds_may15 = [r.id for r in cds_may15_records]
cds_may15[0]
cds_may15[0], len(cds_may15), len(set(cds_may15))
```

```python
with open('./tgd2024_may15/Tetrahymena_Genome_annotation_V2024_mRNA.fasta', 'r') as f:
    mrna_may15_records = list(SeqIO.parse(f, 'fasta'))

mrna_may15 = [r.id for r in mrna_may15_records]
mrna_may15[0], len(mrna_may15), len(set(mrna_may15))
```

### THE DIFFERENCE BETWEEN THE EXON AND CDS FILES

```python
len(set.difference(set(mrna_may15), set(cds_may15))), len(set(mrna_may15)), len(set(cds_may15))
```

The mrna file contains twice as many genes.

```python
set.difference(set(mrna_may15), set(cds_may15))
```

All of the duplicated genes are gids. I wonder if all of the sequences are duplicated.

```python
mrna_may15_g_seqs = {r.id: str(r.seq) for r in mrna_may15_records if 'g' in r.id}
# mrna_may15_ttherm_seqs = {r.id: str(r.seq) for r in mrna_may15_records if 'TTHERM_' in r.id}
# mrna_may15_ttherm_seqs = {r.id: str(r.seq.complement()) for r in mrna_may15_records if 'TTHERM_' in r.id}
# mrna_may15_ttherm_seqs = {r.id: str(r.seq)[::-1] for r in mrna_may15_records if 'TTHERM_' in r.id}
mrna_may15_ttherm_seqs = {r.id: str(r.seq.reverse_complement()) for r in mrna_may15_records if 'TTHERM_' in r.id}
```

```python
len(mrna_may15_g_seqs), list(mrna_may15_g_seqs.keys())[0], len(mrna_may15_ttherm_seqs), list(mrna_may15_ttherm_seqs.keys())[0]
```

```python
sorted(list(mrna_may15_g_seqs.values()))[0]
```

```python
sorted(list(mrna_may15_ttherm_seqs.values()))[0]
```

```python
len(sorted(list(mrna_may15_g_seqs.values()))[0]), len(sorted(list(mrna_may15_ttherm_seqs.values()))[0])
```

```python
sorted(list(mrna_may15_g_seqs.values()))[0] == sorted(list(mrna_may15_ttherm_seqs.values()))[0]
```

```python
not_equal = 0

for idx, id in tqdm.tqdm(enumerate(sorted(list(mrna_may15_g_seqs.values())))):
    if id != sorted(list(mrna_may15_ttherm_seqs.values()))[idx]:
        not_equal += 1

not_equal
```

The sequences are not duplicated.
